```python
import pandas as pd
```

```python
file = "tracks.csv"
tracks_df = pd.read_csv(file)
tracks_df
```

```python
tracks_1 = tracks_df.drop(["id", "id_artists"], axis=1)
tracks_1
```

```python
tracks_modern = tracks_1.loc[tracks_1["release_date"] > "2014-12-31"]
tracks_modern
```

```python
tracks_modern_1 = tracks_modern.drop_duplicates()
tracks_modern_1
```

```python
tracks_modern_2 = tracks_modern_1.drop(["key", "mode", "time_signature"], axis=1)
tracks_modern_2
```

```python
tracks_modern_2.to_csv("tracks_clean.csv")
```
