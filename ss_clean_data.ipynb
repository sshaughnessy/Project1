```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = "tracks_clean.csv"

tracks_df = pd.read_csv(file) #parse_dates=True
tracks_df.head()
```

```python
tracks_df["lower"] = tracks_df["name"].str.lower()
tracks_df.head()
```

```python
remasters = tracks_df[tracks_df["lower"].str.contains("remaster")]
remasters
```

```python
new_df = tracks_df[tracks_df["lower"].str.contains("remaster") == False]
```

```python
new_df
```

```python
output_path = "new_data.csv"
new_df.to_csv(output_path, index=False, header=True)
```
